```python
#importing all libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
#loading iris dataset
iris=datasets.load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
#finding optimum number of clusters(k-means)
x=df.iloc[:,[0,1,2,3]].values
c=[]
for i in range(1,11):
    k=KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
    k.fit(x)
    c.append(k.inertia_)
    
```

```python
#plotting result onto line graph
plt.plot(range(1,11),c)
plt.title("Elbow Graph")
plt.xlabel("Number Of Clusters")
plt.ylabel("Within Cluster Sum Of Squares")
plt.show()
```

```python
#applying k-means to dataset, where number of clusters is determined by graph, which is 3.
k=KMeans(n_clusters=3,init="k-means++",max_iter=300,n_init=10,random_state=0)
y_pred=k.fit_predict(x)
```

```python
#visualising clusters on first to columns
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=100,c="blue",label="SETOSA")
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=100,c="red",label="VERSICOLOUR")
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=100,c="green",label="VIRGINICA")
plt.scatter(k.cluster_centers_[:,0],k.cluster_centers_[:,1],s=100,c="yellow",label="CENTROIDS")
plt.legend(loc="upper right")
```
